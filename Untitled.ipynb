```python
def funcao_ola(x = str):
  print(x)
  trivial = x + " mundo"
  return trivial

trivial = "ola"
resultado = funcao_ola(trivial)
print(resultado)
resultado2 = funcao_ola(resultado)
print(resultado2)
```

```python
def calcula_imc(peso, altura):
    print("Parâmetro peso:, peso")
    print("Parâmetro altura:, altura")
    imc = peso/altura**2
    print(f"IMC: {imc}")

# Início do Programa
print("Início do programa")
calcula_imc(84, 1.80)
print("Término do programa")
```
